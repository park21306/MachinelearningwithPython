```python
import numpy as np

from sklearn.linear_model import LinearRegression,Ridge,RidgeCV,Lasso
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.datasets import make_regression
```

13.1 : Fitting a Line

```python
features,target = make_regression(n_samples=100,
                                  n_features=3,
                                  n_informative=2,
                                  n_targets=1,
                                  noise=0.2,
                                  coef=False,
                                  random_state=1)

regression = LinearRegression()
model = regression.fit(features,target)
```

Discussion

```python
model.intercept_
```

```python
model.coef_
```

```python
target[0]
```

```python
model.predict(features)[0]
```

```python
print(model.score(features,target))
```

13.2 : Handing interactive Effects

```python
features,target = make_regression(n_samples=100,
                                  n_features=2,
                                  n_informative=2,
                                  n_targets=2,
                                  noise=0.2,
                                  coef=False,
                                  random_state=1)

interactive = PolynomialFeatures(degree=3,
                                 include_bias=False,
                                 interaction_only=True)
features_interaction = interactive.fit_transform(features)
regression = LinearRegression()
model = regression.fit(features_interaction,target)
```

Discussion

```python
features[0]
```

```python
interactive_tern = np.multiply(features[:,0],features[:,1])
interactive_tern[0]
```

```python
features_interaction[0]
```

13.3 : Fitting a Nonlinear Relationship

```python
features,target = make_regression(n_samples=100,
                                  n_features=3,
                                  n_informative=2,
                                  n_targets=1,
                                  noise=0.2,
                                  coef=False,
                                  random_state=1)

polynomial = PolynomialFeatures(degree=3,include_bias=False)
features_polynomial = polynomial.fit_transform(features)
regression = LinearRegression()
model = regression.fit(features_polynomial,target)
```

13.4 : Reducing Variance with Regularization

```python
features,target = make_regression(n_samples=100,
                                  n_features=3,
                                  n_informative=2,
                                  n_targets=1,
                                  noise=0.2,
                                  coef=False,
                                  random_state=1)

scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)
regression = Ridge(alpha=0.5)
model = regression.fit(features_standardized,target)
```

Discussion

```python
regr_cv = RidgeCV(alphas=[0.1,1.0,10.0])
model_cv = regr_cv.fit(features_standardized,target)
model_cv.coef_
```

```python
model_cv.alpha_
```

13.5 : Reducing Features with Lasso Regression

```python
features,target = make_regression(n_samples=100,
                                  n_features=3,
                                  n_informative=2,
                                  n_targets=1,
                                  noise=0.2,
                                  coef=False,
                                  random_state=1)

scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)
regression = Lasso(alpha=0.5)
model = regression.fit(features_standardized,target)
```

Discussion

```python
model.coef_
```

```python
regression_a10 = Lasso(alpha=10)
model_a10 = regression_a10.fit(features_standardized,target)
model_a10.coef_
```
