```python
import numpy as np

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans,MiniBatchKMeans,MeanShift,DBSCAN,AgglomerativeClustering
```

19.1 : Clustering Using K-Means

```python
iris = load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = KMeans(n_clusters=3,random_state=0,n_init="auto")

model = cluster.fit(features_std)
```

Discussion

```python
model.labels_
```

```python
iris.target
```

```python
new_observation = [[0.8,0.8,0.8,0.8]]

model.predict(new_observation)
```

```python
model.cluster_centers_
```

19.2 : Speeding Up K-Means Clustering

```python
iris = load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = MiniBatchKMeans(n_clusters=3,
                 random_state=0,
                 batch_size=100,
                 n_init="auto")

model = cluster.fit(features_std)
```

19.3 : Clustering Using Mean Shift

```python
iris = load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = MeanShift(n_jobs=-1)

model = cluster.fit(features_std)
```

19.4 Clustering Using DBSCAN

```python
iris = load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = DBSCAN(n_jobs=-1)

model = cluster.fit(features_std)
```

Discussion

```python
model.labels_
```

19.5 : Clustering Using Hierarchical Merging

```python
iris = load_iris()
features = iris.data

scaler = StandardScaler()
features_std = scaler.fit_transform(features)

cluster = AgglomerativeClustering(n_clusters=3)

model = cluster.fit(features_std)
```

Discussion

```python
model.labels_
```
